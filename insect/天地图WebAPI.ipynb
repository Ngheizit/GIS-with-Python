## 1. POI搜索

```python
# 普通搜索
def CreateJSON_GeneralSearch(keyWord, level, mapBound, queryType, count, start):
    return {
        "keyWord": keyWord,
        "level": level,
        "mapBound": mapBound,
        "queryType": queryType,
        "count": count,
        "start": start
    }
```
